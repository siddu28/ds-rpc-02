```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def preprocess(text):

    doc = nlp(text)
    filtered_text = []
    for token in doc:
        if not token.is_stop and not token.is_punct and not token.like_num:
            filtered_text.append(token.lemma_.lower().strip())
    
    return " ".join(filtered_text)
```

```python
df_2020.info()
```

```python
df_2020["SYMPTOM_TEXT"] = df_2020["SYMPTOM_TEXT"].apply(preprocess)
```

### Preprocessing For 2021 DataSet

```python
df_2021_vares = pd.read_csv(
    r"C:\Users\Siddu\Downloads\New folder\ds-rpc-02\data\raw\2021VAERSDATA.csv",
    encoding="latin1"
)

df_2021_symptoms = pd.read_csv(
    r"C:\Users\Siddu\Downloads\New folder\ds-rpc-02\data\raw\2021VAERSSYMPTOMS.csv",
    encoding="latin1"
)

df_2021_varisvax = pd.read_csv(
    r"C:\Users\Siddu\Downloads\New folder\ds-rpc-02\data\raw\2021VAERSVAX.csv",
    encoding="latin1"
)

cols_to_keep = [
    "VAERS_ID", "AGE_YRS", "SEX", 
    "DIED", "L_THREAT", "HOSPITAL", "DISABLE", 
    "SYMPTOM_TEXT", "SYMPTOM1", "SYMPTOM2", "SYMPTOM3", "SYMPTOM4", "SYMPTOM5",
    "VAX_TYPE", "VAX_MANU", "VAX_NAME"
]

df_2021 = df_2021_vares.merge(df_2021_symptoms,on="VAERS_ID",how="inner")
df_2021 = df_2021.merge(df_2021_varisvax,on="VAERS_ID",how="inner")

df_2021 = df_2021[cols_to_keep]

df_2021 = df_2021.dropna(subset=["SYMPTOM_TEXT"])

df_2021["AGE_YRS"] = df_2021["AGE_YRS"].fillna(-1)
df_2021["SEX"] = df_2021["SEX"].fillna("U")

df_2021["SYMPTOM_TEXT"] = df_2021["SYMPTOM_TEXT"].str.lower().str.strip()


for col in ["DIED", "L_THREAT", "HOSPITAL", "DISABLE"]:
    df_2021[col] = df_2021[col].fillna("N")

for col in ["SYMPTOM2", "SYMPTOM3", "SYMPTOM4", "SYMPTOM5"]:
    df_2021[col] = df_2021[col].fillna("")
```
