```python
import os
RAW_DATA_PATH = r"C:\Users\Siddu\Downloads\New folder\ds-rpc-02\data\raw"
PROCESSED_PATH = r"C:\Users\Siddu\Downloads\New folder\ds-rpc-02\data\processed"
OUTPUT_FILE = os.path.join(PROCESSED_PATH, "master_vaers_data.csv")
```

```python
import glob
def load_and_concat(file_pattern):

    all_files = glob.glob(file_pattern)
    if not all_files:
        print(f"Warning: No files found for pattern {file_pattern}")
        return pd.DataFrame()
        
    print(f"Loading {len(all_files)} files for pattern: {file_pattern}")
    
    df_list = []
    for f in all_files:
        try:
            df = pd.read_csv(f, encoding='latin1', low_memory=False)
            df_list.append(df)
        except Exception as e:
            print(f"Error loading {f}: {e}")
            
    if not df_list:
        print(f"Error: Could not load any files for {file_pattern}")
        return pd.DataFrame()
        
    return pd.concat(df_list, ignore_index=True)
```

```python
# Load all ...DATA.csv files
data_pattern = os.path.join(RAW_DATA_PATH, "*VAERSDATA.csv")
all_data_df = load_and_concat(data_pattern)

# Load all ...SYMPTOMS.csv files
symptoms_pattern = os.path.join(RAW_DATA_PATH, "*VAERSSYMPTOMS.csv")
all_symptoms_df = load_and_concat(symptoms_pattern)

# Load all ...VAX.csv files
vax_pattern = os.path.join(RAW_DATA_PATH, "*VAERSVAX.csv")
all_vax_df = load_and_concat(vax_pattern)
```

```python
all_data_df.shape,all_symptoms_df.shape,all_vax_df.shape
```

```python
symptom_cols = ['SYMPTOM1', 'SYMPTOM2', 'SYMPTOM3', 'SYMPTOM4', 'SYMPTOM5']
```

```python
present_symptom_cols = [col for col in symptom_cols if col in all_symptoms_df.columns]
present_symptom_cols
```

```python
symptoms_melted = all_symptoms_df.melt(
    id_vars=['VAERS_ID'],
    value_vars=present_symptom_cols,
    value_name='structured_symptom'
)
```

```python
symptoms_melted.head()
```

```python
symptoms_melted.shape
```

```python
symptoms_melted.isnull().sum()
```

```python
symptoms_melted = symptoms_melted.dropna(subset=['structured_symptom'])
symptoms_melted.head()
```

```python
agg_symptoms_df = symptoms_melted.groupby('VAERS_ID')['structured_symptom'].apply(list).reset_index()
agg_symptoms_df
```

```python
agg_symptoms_df.rename(columns={'structured_symptom': 'structured_symptoms_list'}, inplace=True)
```

```python
agg_symptoms_df
```

```python
all_vax_df.head()
```

```python
print("Processing vaccine data...")
agg_vax_df = all_vax_df.groupby('VAERS_ID').agg(
    vax_names=('VAX_NAME', list),
    vax_manus=('VAX_MANU', list)
).reset_index()
```

```python
agg_vax_df
```

```python
print("Merging all data sources...")
master_df = all_data_df

# Merge symptoms (how='left' keeps all reports, even if they have no symptoms)
master_df = pd.merge(master_df, agg_symptoms_df, on='VAERS_ID', how='left')

master_df = pd.merge(master_df, agg_vax_df, on='VAERS_ID', how='left')

print("Merge complete.")
```

```python
master_df.head()
```

```python
master_df.shape
```

```python
final_columns = [
    'VAERS_ID',
    'SYMPTOM_TEXT',
    'AGE_YRS',
    'SEX',
    'structured_symptoms_list',
    'vax_names',                
    'vax_manus'                 
]
```

```python
final_columns = [col for col in final_columns if col in master_df.columns]
final_df = master_df[final_columns]
```

```python
final_df.head()
```

```python
for list_col in ['structured_symptoms_list', 'vax_names', 'vax_manus']:
    if list_col in final_df.columns:
        final_df[list_col] = final_df[list_col].apply(lambda x: x if isinstance(x, list) else [])
```

```python
final_df.shape
```

```python
final_df = final_df.dropna(subset=['SYMPTOM_TEXT'])
```

```python
final_df.to_csv(OUTPUT_FILE, index=False)
```
